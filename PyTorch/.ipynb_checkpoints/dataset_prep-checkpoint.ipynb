```python

for i in range(1,26):
    day = pad_day(i)
    if(path.exists("datasets/AAPL/AAPL_daily" + str(i) + ".json")):
        p = pandas.read_json(path_or_buf = "datasets/AAPL/AAPL_daily" + str(i) + ".json")
        if not p.empty:
            prices = p["prices"]
            sell_opps = []
            for j in range(len(prices)):
                subset = prices[j:]
                sell_opps.append(sum(subset - prices[j])/200)


            sell_opps = pandas.DataFrame({'prices': np.clip(sell_opps, 0, 1)})
            minutes = p["minute"]
            df = pandas.concat([minutes, sell_opps], axis=1)

            df.to_json('datasets/AAPL_answer/AAPL_daily_answer' + str(i) + '.json')
```
